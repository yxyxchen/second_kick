```python
conditions, delayMaxs, blockMin, blockSec, nBlock, tokenValue, smallReward, iti, pareto, tGrid, kmGrid = expParas()
```

```python
# constants 
timeWaited_ = np.array([10, 12, 13, 14, 15]);
trialEarnings_ = np.array([10, 10, 10, 10, 10]); 
nTrial = timeWaited_.shape[0]

glr = 0.01
llr = 0.1

globalRate = 0
globalRate_ = []
localRate = 0
localRate_ = []
taskTime_ = []


taskTime = 0

for tIdx in range(nTrial):
    
    t = 0
    while t <= 40:
        if t > timeWaited_[tIdx]:
            globalRate = globalRate + (trialEarnings_[tIdx] - globalRate) * glr
            localRate = localRate + (trialEarnings_[tIdx] - localRate) * llr 
            globalRate_.append(globalRate)
            localRate_.append(localRate)
            taskTime_.append(taskTime)
            break
        
        globalRate = globalRate + (0 - globalRate) * glr
        localRate = localRate + (0 - localRate) * llr 
        
        globalRate_.append(globalRate)
        localRate_.append(localRate)
        
        taskTime_.append(taskTime)
        
        # update
        t += 0.1
        taskTime += 0.1
    
```

```python
# constants 
# parameters 
ltimeconst = 10
gtimeconst = 100 

lprior = tokenValue # prior local reward rate 
gprior = (0.8333333333333335 + 0.9309250282337017)/2 * 0.8 # prior global reward rate 


dr = 0.01 # same drift rates for both actions s
sigma = 0.5 # decision noise 
desThresh = 1 # decision threshold 


# trial-wise variables 
## those without prior values can be local variables 


# timestep-wise recording variables

# trial-wise recording variables 
nPLT = 10 # the number of preloaded trials 
scheduledDelay_ = np.zeros(nPLT)
trialEarnings_ = np.zeros(nPLT)
totalEarnings_ = np.zeros(nPLT)
timeWaited_ = np.zeros(nPLT)
trialStartTime_ = np.zeros(nPLT)

# update throughout time steps 
## value layer is updated in every time second 
## decision layer is not updated during iti 


# iterations 

# cover the entire time line 
nTimeStep = 101
timeSteps, bin = np.linspace(0, delayMaxs[0], num = nTimeStep, endpoint = True, retstep = True)
localRate_ = []
globalRate_ = []
leaveDes_ = []
taskTime_ = []


# variables reset at the very beginning of a task 
localRate = lprior
globalRate = gprior
taskTime = 0
taskTimeIdx = 0
leaveDes = 0


# over trials 
for tIdx in range(nPLT):
    scheduledDelay = max(drawSample("HP"), max(timeSteps)) # just to make sure 
    scheduledDelay_[tIdx] =  scheduledDelay
    trialStartTime_[tIdx] = taskTime
    
    for t in timeSteps:
        if t >= scheduledDelay:
            trialEarnings_[tIdx] = tokenValue
            timeWaited_[tIdx] = t # temporal resolution limits 
    
            # update
            taskTime += iti 
            taskTime_.append(taskTime)
            
            leaveDes = 0
            localRate = localRate + alpha (t)
            globalRate = globalRate +(-localRate + tokenValue) * bin / gtimeconst
            localRate = localRate -localRate * bin / ltimeconst
            globalRate = globalRate -localRate * bin / gtimeconst
            
            localRate_.append(localRate)
            globalRate_.append(globalRate)
            leaveDes_.append(leaveDes)
            break 
            
        # update rewardRates (continous time)
        localRate = localRate - localRate * bin / ltimeconst
        globalRate = globalRate - globalRate * bin / gtimeconst
        localRate_.append(localRate)
        globalRate_.append(globalRate)
        
        # update decisions 
        leaveDes = leaveDes + (dr * (globalRate - localRate) + sigma * np.random.normal(1)) * bin 
        leaveDes_.append(leaveDes)
        taskTime_.append(taskTime)
        
        # update
        taskTime += bin
        
        # check whether to terminate the trial 
        if leaveDes >= desThresh:
            trialEarnings_[tIdx] = 0
            timeWaited_[tIdx] = t

            
            taskTime += iti 
            taskTime_.append(taskTime)
            
            leaveDes = 0
            localRate = localRate  -localRate * bin / ltimeconst
            globalRate = globalRate -localRate * bin / gtimeconst
            localRate = localRate -localRate * bin / ltimeconst
            globalRate = globalRate -localRate * bin / gtimeconst
            
            localRate_.append(localRate)
            globalRate_.append(globalRate)
            leaveDes_.append(leaveDes)
            break
    
        
    
        
        
        
        
    
    
```
